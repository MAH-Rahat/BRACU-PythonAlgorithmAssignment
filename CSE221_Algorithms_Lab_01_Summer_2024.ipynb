## **TASK 01 (a)** image.png

```python
workfile = open("input1a.txt", "r")
newfile = workfile.readlines()
update = []
for i in range (1, len(newfile)):
  s = newfile[i].strip()
  if int(s)%2 == 0:
    update.append(f"{s} is an Even number")
  else:
    update.append(f"{s} is an Odd number")

with open("output1a.txt", 'w') as file:
    for i in update:
        file.write(str(i) + '\n')
```

```python
workfile = open("input1b.txt", "r")

newfile = workfile.readlines()
results = []
for i in range(1, len(newfile)):
  s = newfile[i].strip().split(" ")
  if s[2] == "+":
    result = int(s[1]) + int(s[3])
    results.append(f"The result of {s[1]} {s[2]} {s[3]} is {result}")

  if s[2] == "-":
    result = int(s[1]) - int(s[3])
    results.append(f"The result of {s[1]} {s[2]} {s[3]} is {result}")

  if s[2] == "*":
    result = int(s[1]) * int(s[3])
    results.append(f"The result of {s[1]} {s[2]} {s[3]} is {result}")

  if s[2] == "/":
    result = int(s[1]) / int(s[3])
    results.append(f"The result of {s[1]} {s[2]} {s[3]} is {result}")


with open("output1b.txt", "w") as file:
  for i in results:
    file.write(i + '\n')
```

## **TASK 02**

```python
def bubbleSort(arr):
    n = len(arr)
    for i in range(n - 1):
        swapped = False  # Initialize swapped flag
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:  # Compare adjacent elements
                arr[j], arr[j + 1] = arr[j + 1], arr[j]  # Swap if needed
                swapped = True  # Set swapped to True if a swap is made
        if not swapped:  # If no swaps were made, the array is sorted
            break

with open("input2.txt", "r") as file:
    data = file.read().strip()
    # Convert input data into a list of integers
    arr = list(map(int, data.split()))

bubbleSort(arr)

# Write the sorted output to output2.txt
with open("output2.txt", "w") as file:
    file.write(" ".join(map(str, arr)))

```
